```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
file_path = '../input/movies-on-netflix-prime-video-hulu-and-disney/MoviesOnStreamingPlatforms_updated.csv'

# loading data set
movies_data = pd.read_csv(file_path,index_col=0)
# displaying first five rows
movies_data.head()
```

```python
movies_data.describe()
```

```python
movies_data.info()
```

```python
# scatter plot showing movie year and runtime duration 
sb.scatterplot(x=movies_data.Year,y=movies_data.Runtime,hue=movies_data['IMDb'])
```

```python
#kernal density estimation showing movie release year
sb.kdeplot(data=movies_data.Year,shade=True)
```

```python
# joint plot showing movie data trends with kde
sb.jointplot(x=movies_data.Year,y=movies_data.IMDb,kind='kde')
```
